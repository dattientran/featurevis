```python
import torch
from torch import nn
from featurevis import utils

torch.pi = torch.acos(torch.zeros(1)).item() * 2 # which is 3.1415927410125732

class CombPlaidGenerator(nn.Module):

    def __init__(self, device='cpu', image_size=(36, 64), center=(0., 0.), mask_sigma=5):
        """
        Combined Plaid generator class.
        Params:
            device: 'cuda' or None
            image_size (tuple of integers): Image height and width.
            center (tuple of integers): The center position of the pattern.
            edge_theta (float): Orientation of the boundary between the two plaids (in radian).
            theta (float): Orientation of the sinusoid (in radian).
            sigma (float): std deviation of the Gaussian.
            Lambda (float): Sinusoid wavelengh (1/frequency).
            psi (float): Phase of the sinusoid.
            gamma (float): The ratio between sigma in x-dim over sigma in y-dim (acts
                like an aspect ratio of the Gaussian).

        Returns:
            2D torch.tensor: An image with two combined plaids.
        """
        
        super().__init__()
        self.device = device
        self.image_size = image_size
        self.center = center
        self.mask_sigma = mask_sigma
    
    def gen_gabor(self, theta, Lambda, sigma, psi, gamma):
        ymax, xmax = self.image_size
        xmax, ymax = (xmax - 1)/2, (ymax - 1)/2
        xmin = -xmax
        ymin = -ymax
        (y, x) = torch.meshgrid(torch.arange(ymin, ymax+1), torch.arange(xmin, xmax+1))
        x = x.to(self.device)
        y = y.to(self.device)
        
        # Clip values in reasonable range
        theta.data.clamp_(0, torch.pi)
        psi.data.clamp_(0, torch.pi*2)
        sigma.data.clamp_(3., min(self.image_size)/2) #min(self.image_size)/7, min(self.image_size)/5) #2)

        sigma_x = sigma
        sigma_y = sigma / gamma
        
        # Rotation
        x_theta = (x - self.center[0]) * torch.cos(theta) + (y - self.center[1]) * torch.sin(theta)
        y_theta = -(x - self.center[0]) * torch.sin(theta) + (y - self.center[1]) * torch.cos(theta)
        gb = torch.exp(-.5 * (x_theta ** 2 / sigma_x ** 2 + y_theta ** 2 / sigma_y ** 2)) * torch.cos(2 * torch.pi / Lambda * x_theta + psi)

        return gb.view(1, 1, *self.image_size)
    
    def gen_plaid(self, gabor1, gabor2):
        return gabor1 + gabor2
    
    def gen_comb_plaid(self, plaid1, plaid2, edge_center_x, edge_center_y, edge_theta):

        edge_theta.data.clamp_(-torch.pi/2, torch.pi/2)
        
        ymax, xmax = self.image_size
        xmax, ymax = (xmax - 1)/2, (ymax - 1)/2
        xmin = -xmax
        ymin = -ymax
        (y, x) = torch.meshgrid(torch.arange(ymin, ymax+1), torch.arange(xmin, xmax+1))
        x = x.to(self.device)
        y = y.to(self.device)
        
        px, py = edge_center_x, edge_center_y
        del_x = torch.tensor([1.], dtype=torch.float32).to(self.device)
        qx, qy = (px + del_x, py - del_x * torch.tan(edge_theta))

        mask1 = torch.as_tensor((qx - px) * (y - py) - (qy - py) * (x - px), dtype=torch.float32, device=self.device)
        mask1 = torch.sigmoid(mask1)
        mask2 = torch.ones_like(mask1) - mask1
        
        blur = ops.GaussianBlur(self.mask_sigma)
        blur_mask1 = blur(mask1[None, None])
        blur_mask2 = blur(mask2[None, None])

        return plaid1 * blur_mask1 + plaid2 * blur_mask2, blur_mask1, blur_mask2, mask1, mask2
    
    @utils.varargin
    def forward(self, params):
        edge_center_x, edge_center_y, edge_theta, theta1, angle1, Lambda1, sigma1, psi1, gamma1, theta2, angle2, Lambda2, sigma2, psi2, gamma2 = params
        gb1 = self.gen_gabor(theta1, Lambda1, sigma1, psi1, gamma1)
        gb2 = self.gen_gabor(theta1+angle1, Lambda1, sigma1, psi1, gamma1)
        plaid1 = self.gen_plaid(gb1, gb2)
        gb3 = self.gen_gabor(theta2, Lambda2, sigma2, psi2, gamma2)
        gb4 = self.gen_gabor(theta2+angle2, Lambda2, sigma2, psi2, gamma2)
        plaid2 = self.gen_plaid(gb3, gb4)
        image, blur_mask1, blur_mask2, mask1, mask2 = self.gen_comb_plaid(plaid1, plaid2, edge_center_x, edge_center_y, edge_theta)
        
        return image, blur_mask1, blur_mask2, plaid1, plaid2, gb1, gb2, gb3, gb4,#, mask1, mask2
            
#         return image
```

```python
def center_and_crop_image(image, x_offset, y_offset, output_size=(32, 32)):
    """ Center image to the closest integer and crop it

    Arguments:
        image (array): Original image
        x_offset (float): How far to the right (in x) is the image center.
        y_offset (float): How far to the bottom (in y) is the image center.
        output_size (tuple): How far to the right is the image center.

    Returns:
        Centered image after cropping to the desired output size.
    """
    
    if output_size == (64, 64):
        x_offset = x_offset*2
        y_offset = y_offset*2
    top = (image.shape[0] - output_size[0]) / 2 + y_offset
    bottom = (image.shape[0] - output_size[0]) / 2 - y_offset
    left = (image.shape[1] - output_size[1]) / 2 + x_offset
    right = (image.shape[1] - output_size[1]) / 2 - x_offset
    top, bottom, left, right = (int(round(x)) for x in [top, bottom, left, right])

    # Pad image
    pad_amount = ((abs(top) if top < 0 else 0, abs(bottom) if bottom < 0 else 0),
                  (abs(left) if left < 0 else 0, abs(right) if right < 0 else 0))
    padded = np.pad(image, pad_amount, mode='edge')
    top, bottom, left, right = (np.clip(top, 0, None), np.clip(bottom, 0, None),
                                np.clip(left, 0, None), np.clip(right, 0, None))

    # Crop
    cropped = padded[top:padded.shape[0] - bottom, left:padded.shape[1] - right]
    if cropped.shape != output_size:
        raise ValueError('Something wrong with the cropping. This may fail for '
                         'images with odd dimensions.')

    return cropped

def make_crop_mei(mei, mask_x, mask_y, output_size=(32, 32)):
    top = (mei.shape[0] - output_size[0]) / 2 + mask_y
    bottom = (mei.shape[0] - output_size[0]) / 2 - mask_y
    left = (mei.shape[1] - output_size[1]) / 2 + mask_x
    right = (mei.shape[1] - output_size[1]) / 2 - mask_x
    top, bottom, left, right = (int(round(x)) for x in [top, bottom, left, right])

    # Pad image
    pad_amount = ((abs(top) if top < 0 else 0, abs(bottom) if bottom < 0 else 0),
                  (abs(left) if left < 0 else 0, abs(right) if right < 0 else 0))
    padded = np.pad(mei, pad_amount, mode='edge')
    top, bottom, left, right = (np.clip(top, 0, None), np.clip(bottom, 0, None),
                                np.clip(left, 0, None), np.clip(right, 0, None))

    # Crop
    cropped = padded[top:padded.shape[0] - bottom, left:padded.shape[1] - right]
    if cropped.shape != output_size:
        raise ValueError('Something wrong with the cropping. This may fail for images'
                         'with odd dimensions.')
    return cropped

# centered_mei = center_and_crop_image(mei, x_offset, y_offset)
```

```python
def rotate_mei(mei, mask, mask_x, mask_y, angle, target_std, target_mean):
    nonmask = np.ones(mask.shape) - mask
    background = (mei * nonmask).sum(axis=(-1, -2), keepdims=True) / nonmask.sum()
    h, w = mei.shape
    pad_h, pad_w = int(h/2), int(w/2)
    pad_mei = np.pad(mei, ((pad_h, pad_h), (pad_w, pad_w)), 'constant', constant_values=background)
    center_mei = np.roll(pad_mei, [-int(round(mask_y)), -int(round(mask_x))], axis=[0, 1])
    rotate_mei = rotate(center_mei, angle, reshape=False)
    shift_back_mei = np.roll(rotate_mei, [int(round(mask_y)), int(round(mask_x))], axis=[0, 1])
    crop_mei = shift_back_mei[pad_h:h+pad_h, pad_w:w+pad_w]
    
    # match contrast within mask
    im_mean = (crop_mei * mask).sum(axis=(-1, -2), keepdims=True) / mask.sum()
    im_std = np.sqrt(np.sum(((crop_mei - im_mean) ** 2) * mask, axis=(-1, -2), keepdims=True) / mask.sum())
    unmasked_im = ((crop_mei - im_mean) / im_std) * target_std + target_mean
    masked_im = unmasked_im * mask + (1 - mask) * target_mean
    
    return masked_im
```

```python
def get_tuning(key, mask, mask_x, mask_y, param, param_bound_name, param_name1, param_name2=None, bounds=None, n_points=100, add_rotation=False, f_ratio=False):
    
    # Get parameters
    if bounds == None:
        bounds = (CombPlaidSearchRange & key).fetch1()
    if param_bound_name == 'psi':
        values = np.linspace(bounds['lower_{}'.format(param_bound_name)], bounds['upper_{}'.format(param_bound_name)], n_points)
    elif param_bound_name == 'lambda':
        values = 2**np.linspace(1, 5, n_points)
    
    if add_rotation:
        rot_values = np.linspace(-90, 90, n_points)
    
    mask_stats_params = (MaskStatsParameters & key).fetch1()
    optcp_params = (OptimizationParameters & key).fetch1()

    # Get predictive model
    model_key = ({'group_id': key['group_id'], 'net_hash': key['net_hash']} if
                 optcp_params['use_avg_model'] else key)
    all_keys = (static_models.Model & model_key & 'seed > 1000').fetch('KEY')
    all_models = [(static_models.Model & mk).load_network() for mk in all_keys]
    mean_eyepos = ([0, 0] if (base.Dataset.TrainStats & key).fetch1('norm_eyepos')
                   else (base.Dataset.TrainStats & key).fetch1('mean_eyepos'))
    mean_eyepos = torch.tensor(mean_eyepos, dtype=torch.float32,
                               device='cuda').unsqueeze(0)
    model = models.Ensemble(all_models, key['readout_key'], eye_pos=mean_eyepos,
                            neuron_idx=key['neuron_id'], device='cuda')

    # Prepare generator
    mask = torch.as_tensor(mask, dtype=torch.float32, device='cuda')
    image_size = torch.as_tensor([optcp_params['height'], optcp_params['width']], device='cuda')
    center = torch.as_tensor([mask_x, mask_y], device='cuda')
    generator = CombPlaidGenerator('cuda', image_size, center, optcp_params['mask_sigma'])
    transform = ops.ChangeMaskStats(float(mask_stats_params['fixed_mask_std']), float(mask_stats_params['fixed_mask_mean']), mask, fix_bg=True)
    gen = lambda p: transform(generator(p)[0])

    param_copy = param.copy()
    images, tunings = [], []
    if f_ratio and param_name2 is not None:
        if not add_rotation:
            for val1, val2 in zip(values, values[::-1]):
                if param['Lambda1'] <= param['Lambda2']:  # always use the low-frequency component as component 1
                    param_copy[param_name1] = val1
                    param_copy[param_name2] = val2
                else:
                    param_copy[param_name2] = val1
                    param_copy[param_name1] = val2
                tensor_param = torch.as_tensor(list(param_copy[0]), dtype=torch.float32, device='cuda')
                image = gen(tensor_param).cpu().detach().squeeze().numpy()
                images.append(image)
                tensor_image = torch.as_tensor(image[None, None], dtype=torch.float32, device='cuda')
                with torch.no_grad():
                    act = model(tensor_image).item()
                tunings.append(act)
        else:
            for rot_val in rot_values:
                for val1, val2 in zip(values, values[::-1]):
                    if param['Lambda1'] <= param['Lambda2']:  # always use the low-frequency component as component 1
                        param_copy[param_name1] = val1
                        param_copy[param_name2] = val2
                    else:
                        param_copy[param_name2] = val1
                        param_copy[param_name1] = val2
                    tensor_param = torch.as_tensor(list(param_copy[0]), dtype=torch.float32, device='cuda')
                    image = gen(tensor_param).cpu().detach().squeeze().numpy()
                    image = rotate_mei(image, mask.cpu().detach().squeeze().numpy(), mask_x, mask_y, rot_val, mask_stats_params['fixed_mask_std'], mask_stats_params['fixed_mask_mean'])
                    images.append(image)
                    tensor_image = torch.as_tensor(image[None, None], dtype=torch.float32, device='cuda')
                    with torch.no_grad():
                        act = model(tensor_image).item()
                        
                    tunings.append(act)
            tunings = np.stack(tunings).reshape(n_points, n_points)
                
    else:
        if param_name2 is not None:
            for val1, val2 in product(values, values):
                if param['Lambda1'] <= param['Lambda2']:  # always use the low-frequency component as component 1
                    param_copy[param_name1] = val1
                    param_copy[param_name2] = val2
                else:
                    param_copy[param_name2] = val1
                    param_copy[param_name1] = val2
                tensor_param = torch.as_tensor(list(param_copy[0]), dtype=torch.float32, device='cuda')
                images.append(gen(tensor_param).cpu().detach().squeeze().numpy())
                with torch.no_grad():
                    act = model(gen(tensor_param)).item()
                tunings.append(act)
            tunings = np.stack(tunings).reshape(n_points, n_points)
        else:
            for val in values:
                param_copy[param_name1] = val
                tensor_param = torch.as_tensor(list(param_copy[0]), dtype=torch.float32, device='cuda')
                images.append(gen(tensor_param).cpu().detach().squeeze().numpy())
                with torch.no_grad():
                    act = model(gen(tensor_param)).item()
                tunings.append(act)
    return values, images, tunings
```

```python
# 1D tunings
cell_types = ['texture', 'high_low', 'curvature']

dic = dict()
for ctype in cell_types:

    all_fratios, all_highsfs, all_lowsfs, all_highphases, all_lowphases = [], [], [], [], []
    keys, masks, xs, ys, meis, mei_acts, optcps, acts, params = (base.MEIMask * zd_deis.MaskFixedMEI.proj('mei', 'activation') * high_low.OptimalCombPlaid & 'search_range = 3' & \
    annotation[ctype]).fetch('KEY', 'mask', 'mask_x', 'mask_y', 'mei', 'activation', 'best_image', 'best_activation', 'best_params')
    
    for key, mei_act, mask, mask_x, mask_y, param in zip(keys, mei_acts, masks, xs, ys, params):
        _, _, fratio_acts = get_tuning(key, mask, mask_x, mask_y, param, 'lambda', 'Lambda1', 'Lambda2', None, 20, False, True)
        if param['Lambda1'].item() < param['Lambda2'].item():
            _, _, highsf_acts = get_tuning(key, mask, mask_x, mask_y, param, 'lambda', 'Lambda1', n_points=20)
            _, _, lowsf_acts = get_tuning(key, mask, mask_x, mask_y, param, 'lambda', 'Lambda2', n_points=20)
            _, _, highphase_acts = get_tuning(key, mask, mask_x, mask_y, param, 'psi', 'psi1', n_points=20)
            _, _, lowphase_acts = get_tuning(key, mask, mask_x, mask_y, param, 'psi', 'psi2', n_points=20)
        else:
            _, _, highsf_acts = get_tuning(key, mask, mask_x, mask_y, param, 'lambda', 'Lambda2', n_points=20)
            _, _, lowsf_acts = get_tuning(key, mask, mask_x, mask_y, param, 'lambda', 'Lambda1', n_points=20)
            _, _, highphase_acts = get_tuning(key, mask, mask_x, mask_y, param, 'psi', 'psi2', n_points=20)
            _, _, lowphase_acts = get_tuning(key, mask, mask_x, mask_y, param, 'psi', 'psi1', n_points=20)

        all_highsfs.append(np.array(highsf_acts) / mei_act)
        all_lowsfs.append(np.array(lowsf_acts) / mei_act)
        all_highphases.append(np.array(highphase_acts) / mei_act)
        all_lowphases.append(np.array(lowphase_acts) /  mei_act)
        all_fratios.append(np.array(fratio_acts) /  mei_act)
        
    dic[ctype] = dict(highsfs=all_highsfs, lowsfs=all_lowsfs, highphases=all_highphases, lowphases=all_lowphases, fratios=all_fratios)
    
```
